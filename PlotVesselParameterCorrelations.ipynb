```python
# -*- coding: utf-8 -*-
"""
Created on Thu Mar 12 23:21:07 2020

@author: whitma01
"""

import csv
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
import pandas as pd
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib.ticker import FuncFormatter
#http://seaborn.pydata.org/index.html
#http://seaborn.pydata.org/tutorial/regression.html#regression-tutorial

sns.set_style("ticks")

data_MISSION = genfromtxt('parameters/measurements_vessel_MISSION_predicted.csv', delimiter=',')
Densities_predicted1 = data_MISSION[1:,1]
AreaFractions_predicted1 = data_MISSION[1:,2]
Median_Area_predicted1 = data_MISSION[1:,3]
Median_Circularity_predicted1 = data_MISSION[1:,4]
Median_MajorMinorAxisRatio_predicted1 = data_MISSION[1:,5]
Median_Thickness_predicted1 = data_MISSION[1:,6]
Mean_Area_predicted1 = data_MISSION[1:,7]
Mean_Circularity_predicted1 = data_MISSION[1:,8]
Mean_MajorMinorAxisRatio_predicted1 = data_MISSION[1:,9]
Mean_Thickness_predicted1 = data_MISSION[1:,10]

data_MISSION = genfromtxt('parameters/measurements_vessel_MISSION_true.csv', delimiter=',')
Densities_true1 = data_MISSION[1:,1]
AreaFractions_true1 = data_MISSION[1:,2]
Median_Area_true1 = data_MISSION[1:,3]
Median_Circularity_true1 = data_MISSION[1:,4]
Median_MajorMinorAxisRatio_true1 = data_MISSION[1:,5]
Median_Thickness_true1 = data_MISSION[1:,6]
Mean_Area_true1 = data_MISSION[1:,7]
Mean_Circularity_true1 = data_MISSION[1:,8]
Mean_MajorMinorAxisRatio_true1 = data_MISSION[1:,9]
Mean_Thickness_true1 = data_MISSION[1:,10]


data_BEHOLD = genfromtxt('parameters/measurements_vessel_BEHOLD_predicted.csv', delimiter=',')
Densities_predicted2 = data_BEHOLD[1:,1]
AreaFractions_predicted2 = data_BEHOLD[1:,2]
Median_Area_predicted2 = data_BEHOLD[1:,3]
Median_Circularity_predicted2 = data_BEHOLD[1:,4]
Median_MajorMinorAxisRatio_predicted2 = data_BEHOLD[1:,5]
Median_Thickness_predicted2 = data_BEHOLD[1:,6]
Mean_Area_predicted2 = data_BEHOLD[1:,7]
Mean_Circularity_predicted2 = data_BEHOLD[1:,8]
Mean_MajorMinorAxisRatio_predicted2 = data_BEHOLD[1:,9]
Mean_Thickness_predicted2 = data_BEHOLD[1:,10]

data_BEHOLD = genfromtxt('parameters/measurements_vessel_BEHOLD_true.csv', delimiter=',')
Densities_true2 = data_BEHOLD[1:,1]
AreaFractions_true2 = data_BEHOLD[1:,2]
Median_Area_true2 = data_BEHOLD[1:,3]
Median_Circularity_true2 = data_BEHOLD[1:,4]
Median_MajorMinorAxisRatio_true2 = data_BEHOLD[1:,5]
Median_Thickness_true2 = data_BEHOLD[1:,6]
Mean_Area_true2 = data_BEHOLD[1:,7]
Mean_Circularity_true2 = data_BEHOLD[1:,8]
Mean_MajorMinorAxisRatio_true2 = data_BEHOLD[1:,9]
Mean_Thickness_true2 = data_BEHOLD[1:,10]


Densities_predicted = np.concatenate((Densities_predicted1, Densities_predicted2), axis=0)
AreaFractions_predicted = np.concatenate((AreaFractions_predicted1, AreaFractions_predicted2), axis=0)
Median_Area_predicted = np.concatenate((Median_Area_predicted1, Median_Area_predicted2), axis=0)
Median_Circularity_predicted = np.concatenate((Median_Circularity_predicted1, Median_Circularity_predicted2), axis=0)
Median_MajorMinorAxisRatio_predicted = np.concatenate((Median_MajorMinorAxisRatio_predicted1, Median_MajorMinorAxisRatio_predicted2), axis=0)
Median_Thickness_predicted = np.concatenate((Median_Thickness_predicted1, Median_Thickness_predicted2), axis=0)
Mean_Area_predicted = np.concatenate((Mean_Area_predicted1, Mean_Area_predicted2), axis=0)
Mean_Circularity_predicted = np.concatenate((Mean_Circularity_predicted1, Mean_Circularity_predicted2), axis=0)
Mean_MajorMinorAxisRatio_predicted = np.concatenate((Mean_MajorMinorAxisRatio_predicted1, Mean_MajorMinorAxisRatio_predicted2), axis=0)
Mean_Thickness_predicted = np.concatenate((Mean_Thickness_predicted1, Mean_Thickness_predicted2), axis=0)

Densities_true = np.concatenate((Densities_true1, Densities_true2), axis=0)
AreaFractions_true = np.concatenate((AreaFractions_true1, AreaFractions_true2), axis=0)
Median_Area_true = np.concatenate((Median_Area_true1, Median_Area_true2), axis=0)
Median_Circularity_true = np.concatenate((Median_Circularity_true1, Median_Circularity_true2), axis=0)
Median_MajorMinorAxisRatio_true = np.concatenate((Median_MajorMinorAxisRatio_true1, Median_MajorMinorAxisRatio_true2), axis=0)
Median_Thickness_true = np.concatenate((Median_Thickness_true1, Median_Thickness_true2), axis=0)
Mean_Area_true = np.concatenate((Mean_Area_true1, Mean_Area_true2), axis=0)
Mean_Circularity_true = np.concatenate((Mean_Circularity_true1, Mean_Circularity_true2), axis=0)
Mean_MajorMinorAxisRatio_true = np.concatenate((Mean_MajorMinorAxisRatio_true1, Mean_MajorMinorAxisRatio_true2), axis=0)
Mean_Thickness_true = np.concatenate((Mean_Thickness_true1, Mean_Thickness_true2), axis=0)



Densities_r, Densities_p = stats.pearsonr(Densities_predicted, Densities_true)
AreaFractions_r, AreaFractions_p = stats.pearsonr(AreaFractions_predicted, AreaFractions_true)
Median_Area_r, Median_Area_p = stats.pearsonr(Median_Area_predicted, Median_Area_true)
Median_Circularity_r, Median_Circularity_p = stats.pearsonr(Median_Circularity_predicted, Median_Circularity_true)
Median_MajorMinorAxisRatio_r, Median_MajorMinorAxisRatio_p = stats.pearsonr(Median_MajorMinorAxisRatio_predicted, Median_MajorMinorAxisRatio_true)
Median_Thickness_r, Median_Thickness_p = stats.pearsonr(Median_Thickness_predicted, Median_Thickness_true)

Mean_Area_r, Mean_Area_p = stats.pearsonr(Mean_Area_predicted, Mean_Area_true)
Mean_Circularity_r, Mean_Circularity_p = stats.pearsonr(Mean_Circularity_predicted, Mean_Circularity_true)
Mean_MajorMinorAxisRatio_r, Mean_MajorMinorAxisRatio_p = stats.pearsonr(Mean_MajorMinorAxisRatio_predicted, Mean_MajorMinorAxisRatio_true)
Mean_Thickness_r, Mean_Thickness_p = stats.pearsonr(Mean_Thickness_predicted, Mean_Thickness_true)



Densities_data1 = {'ground truth':Densities_true1, 'measured':Densities_predicted1}
Densities_frame1 = pd.DataFrame(Densities_data1, columns=['ground truth', 'measured'])
Densities_data2 = {'ground truth':Densities_true2, 'measured':Densities_predicted2}
Densities_frame2 = pd.DataFrame(Densities_data2, columns=['ground truth', 'measured'])
Densities_data = {'ground truth':Densities_true, 'measured':Densities_predicted}
Densities_frame = pd.DataFrame(Densities_data, columns=['ground truth', 'measured'])

AreaFractions_data1 = {'ground truth':AreaFractions_true1, 'measured':AreaFractions_predicted1}
AreaFractions_frame1 = pd.DataFrame(AreaFractions_data1, columns=['ground truth', 'measured'])
AreaFractions_data2 = {'ground truth':AreaFractions_true2, 'measured':AreaFractions_predicted2}
AreaFractions_frame2 = pd.DataFrame(AreaFractions_data2, columns=['ground truth', 'measured'])
AreaFractions_data = {'ground truth':AreaFractions_true, 'measured':AreaFractions_predicted}
AreaFractions_frame = pd.DataFrame(AreaFractions_data, columns=['ground truth', 'measured'])

Median_Area_data1 = {'ground truth':Median_Area_true1, 'measured':Median_Area_predicted1}
Median_Area_frame1 = pd.DataFrame(Median_Area_data1, columns=['ground truth', 'measured'])
Median_Area_data2 = {'ground truth':Median_Area_true2, 'measured':Median_Area_predicted2}
Median_Area_frame2 = pd.DataFrame(Median_Area_data2, columns=['ground truth', 'measured'])
Median_Area_data = {'ground truth':Median_Area_true, 'measured':Median_Area_predicted}
Median_Area_frame = pd.DataFrame(Median_Area_data, columns=['ground truth', 'measured'])

Median_Circularity_data1 = {'ground truth':Median_Circularity_true1, 'measured':Median_Circularity_predicted1}
Median_Circularity_frame1 = pd.DataFrame(Median_Circularity_data1, columns=['ground truth', 'measured'])
Median_Circularity_data2 = {'ground truth':Median_Circularity_true2, 'measured':Median_Circularity_predicted2}
Median_Circularity_frame2 = pd.DataFrame(Median_Circularity_data2, columns=['ground truth', 'measured'])
Median_Circularity_data = {'ground truth':Median_Circularity_true, 'measured':Median_Circularity_predicted}
Median_Circularity_frame = pd.DataFrame(Median_Circularity_data, columns=['ground truth', 'measured'])

Median_MajorMinorAxisRatio_data1 = {'ground truth':Median_MajorMinorAxisRatio_true1, 'measured':Median_MajorMinorAxisRatio_predicted1}
Median_MajorMinorAxisRatio_frame1 = pd.DataFrame(Median_MajorMinorAxisRatio_data1, columns=['ground truth', 'measured'])
Median_MajorMinorAxisRatio_data2 = {'ground truth':Median_MajorMinorAxisRatio_true2, 'measured':Median_MajorMinorAxisRatio_predicted2}
Median_MajorMinorAxisRatio_frame2 = pd.DataFrame(Median_MajorMinorAxisRatio_data2, columns=['ground truth', 'measured'])
Median_MajorMinorAxisRatio_data = {'ground truth':Median_MajorMinorAxisRatio_true, 'measured':Median_MajorMinorAxisRatio_predicted}
Median_MajorMinorAxisRatio_frame = pd.DataFrame(Median_MajorMinorAxisRatio_data, columns=['ground truth', 'measured'])

Median_Thickness_data1 = {'ground truth':Median_Thickness_true1, 'measured':Median_Thickness_predicted1}
Median_Thickness_frame1 = pd.DataFrame(Median_Thickness_data1, columns=['ground truth', 'measured'])
Median_Thickness_data2 = {'ground truth':Median_Thickness_true2, 'measured':Median_Thickness_predicted2}
Median_Thickness_frame2 = pd.DataFrame(Median_Thickness_data2, columns=['ground truth', 'measured'])
Median_Thickness_data = {'ground truth':Median_Thickness_true, 'measured':Median_Thickness_predicted}
Median_Thickness_frame = pd.DataFrame(Median_Thickness_data, columns=['ground truth', 'measured'])


Mean_Area_data1 = {'ground truth':Mean_Area_true1, 'measured':Mean_Area_predicted1}
Mean_Area_frame1 = pd.DataFrame(Mean_Area_data1, columns=['ground truth', 'measured'])
Mean_Area_data2 = {'ground truth':Mean_Area_true2, 'measured':Mean_Area_predicted2}
Mean_Area_frame2 = pd.DataFrame(Mean_Area_data2, columns=['ground truth', 'measured'])
Mean_Area_data = {'ground truth':Mean_Area_true, 'measured':Mean_Area_predicted}
Mean_Area_frame = pd.DataFrame(Mean_Area_data, columns=['ground truth', 'measured'])

Mean_Circularity_data1 = {'ground truth':Mean_Circularity_true1, 'measured':Mean_Circularity_predicted1}
Mean_Circularity_frame1 = pd.DataFrame(Mean_Circularity_data1, columns=['ground truth', 'measured'])
Mean_Circularity_data2 = {'ground truth':Mean_Circularity_true2, 'measured':Mean_Circularity_predicted2}
Mean_Circularity_frame2 = pd.DataFrame(Mean_Circularity_data2, columns=['ground truth', 'measured'])
Mean_Circularity_data = {'ground truth':Mean_Circularity_true, 'measured':Mean_Circularity_predicted}
Mean_Circularity_frame = pd.DataFrame(Mean_Circularity_data, columns=['ground truth', 'measured'])

Mean_MajorMinorAxisRatio_data1 = {'ground truth':Mean_MajorMinorAxisRatio_true1, 'measured':Mean_MajorMinorAxisRatio_predicted1}
Mean_MajorMinorAxisRatio_frame1 = pd.DataFrame(Mean_MajorMinorAxisRatio_data1, columns=['ground truth', 'measured'])
Mean_MajorMinorAxisRatio_data2 = {'ground truth':Mean_MajorMinorAxisRatio_true2, 'measured':Mean_MajorMinorAxisRatio_predicted2}
Mean_MajorMinorAxisRatio_frame2 = pd.DataFrame(Mean_MajorMinorAxisRatio_data2, columns=['ground truth', 'measured'])
Mean_MajorMinorAxisRatio_data = {'ground truth':Mean_MajorMinorAxisRatio_true, 'measured':Mean_MajorMinorAxisRatio_predicted}
Mean_MajorMinorAxisRatio_frame = pd.DataFrame(Mean_MajorMinorAxisRatio_data, columns=['ground truth', 'measured'])

Mean_Thickness_data1 = {'ground truth':Mean_Thickness_true1, 'measured':Mean_Thickness_predicted1}
Mean_Thickness_frame1 = pd.DataFrame(Mean_Thickness_data1, columns=['ground truth', 'measured'])
Mean_Thickness_data2 = {'ground truth':Mean_Thickness_true2, 'measured':Mean_Thickness_predicted2}
Mean_Thickness_frame2 = pd.DataFrame(Mean_Thickness_data2, columns=['ground truth', 'measured'])
Mean_Thickness_data = {'ground truth':Mean_Thickness_true, 'measured':Mean_Thickness_predicted}
Mean_Thickness_frame = pd.DataFrame(Mean_Thickness_data, columns=['ground truth', 'measured'])


f, axes = plt.subplots(3, 2, figsize=(10, 15))

f.subplots_adjust(hspace=0.3, wspace=0.3)
p1=sns.regplot('ground truth', 'measured', Densities_frame, ci=95, ax=axes[0,0], scatter = False , line_kws={"color": "black", 'alpha': 0.5})
p1.set_title('density (#/mm$^2$)', fontsize=18)
p1.set(xlim=(10, 110), ylim=(10, 110))
ymin, ymax = p1.get_ylim()
diffy = ymax - ymin
xmin, xmax = p1.get_xlim()
diffx = xmax - xmin
if Densities_p < 0.001:
    p1.text(xmin + diffx/20, ymax - diffy/6, "r = %5.3f\np < 0.001" %(Densities_r), fontsize=12)
else:
    p1.text(xmin + diffx/20, ymax - diffy/6, "r = %5.3f\np = %5.3f" %(Densities_r, Densities_p), fontsize=12)
p1=sns.regplot('ground truth', 'measured', Densities_frame2, ci=95, ax=axes[0,0], fit_reg = False, scatter_kws={"color": "#01796F", 'alpha': 0.75})
p1=sns.regplot('ground truth', 'measured', Densities_frame1, ci=95, ax=axes[0,0], fit_reg = False, scatter_kws={"color": "#FF69B4", 'alpha': 0.75})
p1.set_xlabel("ground truth",fontsize=14)
p1.set_ylabel("measured",fontsize=14)
for item in ([p1.xaxis.label, p1.yaxis.label] + p1.get_xticklabels() + p1.get_yticklabels()):
    item.set_fontsize(14)

axes[0, 1].axis('off')

p1=sns.regplot('ground truth', 'measured', Mean_Area_frame, ci=95, ax=axes[1,0], scatter = False , line_kws={"color": "black", 'alpha': 0.5})
p1.set_title('mean area (μm$^2$)', fontsize=18)
p1.set(xlim=(0, 1200), ylim=(0, 1200))
ymin, ymax = p1.get_ylim()
diffy = ymax - ymin
xmin, xmax = p1.get_xlim()
diffx = xmax - xmin
if Mean_Area_p < 0.001:
    p1.text(xmin + diffx/20, ymax - diffy/6, "r = %5.3f\np < 0.001" %(Mean_Area_r), fontsize=12)
else:
    p1.text(xmin + diffx/20, ymax - diffy/6, "r = %5.3f\np = %5.3f" %(Mean_Area_r, Mean_Area_p), fontsize=12)
p1=sns.regplot('ground truth', 'measured', Mean_Area_frame2, ci=95, ax=axes[1,0], fit_reg = False, scatter_kws={"color": "#01796F", 'alpha': 0.75})
p1=sns.regplot('ground truth', 'measured', Mean_Area_frame1, ci=95, ax=axes[1,0], fit_reg = False, scatter_kws={"color": "#FF69B4", 'alpha': 0.75})
p1.set_xlabel("ground truth",fontsize=14)
p1.set_ylabel("measured",fontsize=14)
for item in ([p1.xaxis.label, p1.yaxis.label] + p1.get_xticklabels() + p1.get_yticklabels()):
    item.set_fontsize(14)


p1=sns.regplot('ground truth', 'measured', Mean_Circularity_frame, ci=95, ax=axes[1,1], scatter = False , line_kws={"color": "black", 'alpha': 0.5})
p1.set_title('mean circularity', fontsize=18)
p1.set(xlim=(0.45, 0.65), ylim=(0.45, 0.65))
ymin, ymax = p1.get_ylim()
diffy = ymax - ymin
xmin, xmax = p1.get_xlim()
diffx = xmax - xmin
if Mean_Circularity_p < 0.001:
    p1.text(xmin + diffx/20, ymax - diffy/6, "r = %5.3f\np < 0.001" %(Mean_Circularity_r), fontsize=12)
else:
    p1.text(xmin + diffx/20, ymax - diffy/6, "r = %5.3f\np = %5.3f" %(Mean_Circularity_r, Mean_Circularity_p), fontsize=12)
p1=sns.regplot('ground truth', 'measured', Mean_Circularity_frame2, ci=95, ax=axes[1,1], fit_reg = False, scatter_kws={"color": "#01796F", 'alpha': 0.75})
p1=sns.regplot('ground truth', 'measured', Mean_Circularity_frame1, ci=95, ax=axes[1,1], fit_reg = False, scatter_kws={"color": "#FF69B4", 'alpha': 0.75})
p1.set_xlabel("ground truth",fontsize=14)
p1.set_ylabel("measured",fontsize=14)
p1.set(xticks=np.arange(0.45, 0.7, 0.05))
p1.set(yticks=np.arange(0.45, 0.7, 0.05))
for item in ([p1.xaxis.label, p1.yaxis.label] + p1.get_xticklabels() + p1.get_yticklabels()):
    item.set_fontsize(14)

p1=sns.regplot('ground truth', 'measured', Mean_MajorMinorAxisRatio_frame, ci=95, ax=axes[2,0], scatter = False , line_kws={"color": "black", 'alpha': 0.5})
p1.set_title('mean axis ratio', fontsize=18) #, fontweight="bold"
p1.set(xlim=(2, 3.3), ylim=(2, 3.3))
ymin, ymax = p1.get_ylim()
diffy = ymax - ymin
xmin, xmax = p1.get_xlim()
diffx = xmax - xmin
if Mean_MajorMinorAxisRatio_p < 0.001:
    p1.text(xmin + diffx/20, ymax - diffy/6, "r = %5.3f\np < 0.001" %(Mean_MajorMinorAxisRatio_r), fontsize=12)
else:
    p1.text(xmin + diffx/20, ymax - diffy/6, "r = %5.3f\np = %5.3f" %(Mean_MajorMinorAxisRatio_r, Mean_MajorMinorAxisRatio_p), fontsize=12)
p1=sns.regplot('ground truth', 'measured', Mean_MajorMinorAxisRatio_frame2, ci=95, ax=axes[2,0], fit_reg = False, scatter_kws={"color": "#01796F", 'alpha': 0.75})
p1=sns.regplot('ground truth', 'measured', Mean_MajorMinorAxisRatio_frame1, ci=95, ax=axes[2,0], fit_reg = False, scatter_kws={"color": "#FF69B4", 'alpha': 0.75})
p1.set_xlabel("ground truth",fontsize=14)
p1.set_ylabel("measured",fontsize=14)
for item in ([p1.xaxis.label, p1.yaxis.label] + p1.get_xticklabels() + p1.get_yticklabels()):
    item.set_fontsize(14)


p1=sns.regplot('ground truth', 'measured', Mean_Thickness_frame, ci=95, ax=axes[2,1], scatter = False , line_kws={"color": "black", 'alpha': 0.5})
p1.set_title('mean thickness (μm)', fontsize=18)
p1.set(xlim=(6, 12), ylim=(6, 12))
ymin, ymax = p1.get_ylim()
diffy = ymax - ymin
xmin, xmax = p1.get_xlim()
diffx = xmax - xmin
if Mean_Thickness_p < 0.001:
    p1.text(xmin + diffx/20, ymax - diffy/6, "r = %5.3f\np < 0.001" %(Mean_Thickness_r), fontsize=12)
else:
    p1.text(xmin + diffx/20, ymax - diffy/6, "r = %5.3f\np = %5.3f" %(Mean_Thickness_r, Mean_Thickness_p), fontsize=12)
p1=sns.regplot('ground truth', 'measured', Mean_Thickness_frame2, ci=95, ax=axes[2,1], fit_reg = False, scatter_kws={"color": "#01796F", 'alpha': 0.75})
p1=sns.regplot('ground truth', 'measured', Mean_Thickness_frame1, ci=95, ax=axes[2,1], fit_reg = False, scatter_kws={"color": "#FF69B4", 'alpha': 0.75})
p1.set_xlabel("ground truth",fontsize=14)
p1.set_ylabel("measured",fontsize=14)
for item in ([p1.xaxis.label, p1.yaxis.label] + p1.get_xticklabels() + p1.get_yticklabels()):
    item.set_fontsize(14)


x = [0, 0]
y = [0, 0]
classes = ['BEHOLD','MISSION']
values = [0, 1]
colours = ListedColormap(['#01796F','#FF69B4'])
scatter = plt.scatter(x, y,c=values, cmap=colours)
leg = f.legend(handles=scatter.legend_elements()[0], labels=classes, loc='upper center', ncol=1, fancybox=True, shadow=True,fontsize=20)


# Get the bounding box of the original legend
bb = leg.get_bbox_to_anchor().inverse_transformed(p1.transAxes)

# Change to location of the legend. 
bb.y0 += 0.0
bb.y1 -= 0.9
bb.x0 += 0.0
bb.x1 += 1.3
leg.set_bbox_to_anchor(bb, transform = p1.transAxes)


#plt.show()
#fig.savefig("parameters/correlations.pdf", bbox_inches='tight')
plt.savefig("parameters/correlations.pdf", bbox_inches = 'tight')
```
